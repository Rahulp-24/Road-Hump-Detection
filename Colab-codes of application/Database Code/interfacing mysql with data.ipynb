```python
sql='''create table road_t (
source_node_id varchar(20) not null,
destination_node_id varchar(20) not null,
humps_locations varchar(500) ,
pothole_locations varchar(500),
wornout_locations varchar(900),
road_time date ,
pothole date , 
wornout date ,
primary key(source_node_id,destination_node_id))
'''
myCursor.execute(sql)
```

```python
myCursor.execute("SHOW tables")
for i in myCursor:
    print(i)
    
```

```python
myCursor.execute("Truncate table road_t")
```

```python

sql1='''
UPDATE road_t SET humps_locations= %s, road_time=STR_TO_DATE(%s,'%d-%m-%Y')  WHERE source_node_id = %s and destination_node_id=%s
'''
sql2='''
insert into road_t (source_node_id, destination_node_id, humps_locations, road_time) values(%s,%s,%s,%s) 
'''
check='''
select sum(if(source_node_id= %s and destination_node_id=%s ,1,0)) from road_t;
'''
from datetime import date
today = date.today()
d1 = today.strftime("%d-%m-%Y")
d1="02-06-2022"
print(d1)

with open('humps.txt') as f:
    for line in f:      
        l=line.split('->')
        a=l[1].split(' ')
        l.pop()
        str=a[1]        
        l.extend(a[:1]) 
        l.append(str[:-1])
        #print(l)
        val0=(l[0],l[1])
        myCursor.execute(check,val0)
        res=myCursor.fetchall()
        
        res=list(res[0])
        #print(res[0])
        if res[0] is None:
            res= False
        elif res[0]==1:
            res= True
        else:
            res=False
         
        if(res):
            val1=(l[2],d1,l[0],l[1])
            myCursor.execute(sql1, val1)
            #print(val1)
        else:
            val2=(l[0],l[1],l[2],today)
            myCursor.execute(sql2, val2)
            
```
